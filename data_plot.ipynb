```python
import pandas as pd

# Year
year = ['2008-09',
        '2009-10',
        '2010-11',
        '2011-12',
        '2012-13',
        '2013-14',
        '2014-15',
        '2015-16',
        '2016-17',
        '2017-18',
        '2018-19',
        '2019-20',
        '2020-21',
        '2021-22']

# Inpatient Patient Days
inpatient = [7034185,
             7066534,
             7203356,
             7217755,
             7298210,
             7479088,
             7584679,
             7757181,
             7999106,
             8271101,
             8336190,
             8167243,
             7526207,
             7926440
             ]

# Day Inpatient Discharges and Deaths
discharges_deaths = [365222,
                      416885,
                      459548,
                      496640,
                      516217,
                      542333,
                      571563,
                      589623,
                      633508,
                      659413,
                      681985,
                      683477,
                      675649,
                      732499
                     ]

# Patient Days
patient_num = [7399407,
              7483419,
              7662904,
              7714395,
              7814337,
              8021421,
              8156242,
              8346804,
              8632614,
              8930514,
              9018175,
              8850720,
              8201856,
              8658939
              ]

# calculate the number of days in each year
days_in_year = [365, 365, 365, 366, 365, 365, 365, 366, 365, 365, 365, 366, 366, 365]

# calculate daily number of inpatients, discharges and deaths, and patient numbers
daily_inpatient = [round(x/y,2) for x,y in zip(inpatient, days_in_year)]
daily_discharges_deaths = [round(x/y,2) for x,y in zip(discharges_deaths, days_in_year)]
daily_patient_num = [round(x/y,2) for x,y in zip(patient_num, days_in_year)]

# create a new dataframe with the calculated values
data = {'Year': year,
        'Inpatient': inpatient,
        'Discharges & Deaths': discharges_deaths,
        'Patient': patient_num,
        'daily_inpatients': daily_inpatient,
        'daily_discharges_deaths': daily_discharges_deaths,
        'daily_patient_num': daily_patient_num}

df_medical = pd.DataFrame(data)
df_medical
```

```python
df_medical.dtypes
```

```python
colour_map = ['darkgrey','steelblue','skyblue']
ax = df_medical.plot(x="Year", 
                     y=["daily_discharges_deaths", 
                        "daily_inpatients", 
                        "daily_patient_num"], 
                     kind="bar", 
                     color=colour_map, 
                     title='Average Daily Demand for Medical Service, \n \
                        2008-09 to 2021-22',
                     rot=0)
ax.set_xticklabels(df_medical['Year'], rotation=45)
plt.show()
```

```python
df_medical.to_csv('daily_data.csv',index=False)
```
